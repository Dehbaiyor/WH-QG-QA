```python
%%time
import dill as pickle
with open("test_questions/a1_questions.dill", "rb") as dill_file:
    data = pickle.load(dill_file)
print(len(data))

passage_embedding, passages, sentences = get_embs("test_questions/a1.txt")

accuracy = 0
count = 0
for i in range(len(data)):
    try:
        question = data[i]['question']
        ans = get_wh_answer(sentences, question, passage_embedding, passages, 1)
        print("Question ", count + 1,": Pred Ans - ", ans, " Correct Ans - ", data[i]['answer'])
        p = get_position(ans, data[i]['answer'])
        if p != 0:
            accuracy += 1/p
        count += 1
    except:
        pass
    
print(accuracy, " of ", count,": " ,accuracy/count)
```
